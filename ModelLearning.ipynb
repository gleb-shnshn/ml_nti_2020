```python
import pandas as pd
import numpy as np
```

```python
X_train_in = pd.read_csv("X_train_app.csv", index_col=0)
y_train_in = pd.read_csv("y_train.csv",index_col=0)
X_val = pd.read_csv("X_val_app.csv", index_col=0)
#X_test, y_train, y_test 
```

```python
y_train_in["mark"] = (y_train_in["mark"] - 3) / 2
```

```python
X_train_in
```

```python
X_train_in["mark"]=y_train_in["mark"]
X_train_in["STD_ID"]= X_train_in["STD_ID"].astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_in,X_train_in["mark"], shuffle=True, test_size=0.2, random_state=42)
```

```python
y_test.shape, y_train.shape
```

```python
import catboost
```

```python
X_train_in.columns
```

```python
train_col = ['STD_ID_mean','ДИСЦИПЛИНА_mean', 'АТТЕСТАЦИЯ_mean', 'НАПРАВЛЕНИЕ_mean', 'ГОД_mean',
       'КУРС_mean', 'СЕМЕСТР_mean', 'Категория обучения_mean', 'Пол_mean', 'Что именно закончил_mean', 'НАПРАВЛЕНИЕ_clustered_mean']

```

```python
X_train_in
```

```python
params = {
    'n_estimators': 5000,
    'depth': 10,
    'random_state':42,
    'verbose': 100,
    'eval_metric': "R2"
    
}
```

```python
cb = catboost.CatBoostRegressor(**params)
```

```python
cb.fit(X_train[train_col],y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
def app(y):
    return round(y*2 + 3.1)
```

```python
mean_absolute_error(y_test, cb.predict(X_test[train_col])), r2_score(y_test, cb.predict(X_test[train_col]))
```

```python
y_pred = pd.read_csv("sample_submission.csv",index_col=0)
y_pred["mark"] = cb.predict(X_val[train_col])
```

```python
cb.get_feature_importance()
```

```python
submission = round(y_pred*2 + 3)
```
