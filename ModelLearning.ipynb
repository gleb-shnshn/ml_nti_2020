```python
import pandas as pd
import numpy as np
```

```python
X_train_in = pd.read_csv("X_train_app.csv", index_col=0)
y_train_in = pd.read_csv("y_train.csv",index_col=0)
X_val = pd.read_csv("X_val_app.csv", index_col=0)
```

# Нормируем оценку из 3-5 в 0-1

```python
y_train_in["mark"] = (y_train_in["mark"] - 3) / 2
```

```python
X_train_in
```

```python
X_train_in["mark"] = y_train_in["mark"]
X_train_in["STD_ID"]= X_train_in["STD_ID"].astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_in,
                                                    X_train_in["mark"], 
                                                    shuffle=True, 
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
y_test.shape, y_train.shape
```

```python
import catboost
```

```python
X_train_in.columns
```

```python
train_col = ['STD_ID_mean','ДИСЦИПЛИНА_mean', 'АТТЕСТАЦИЯ_mean', 'НАПРАВЛЕНИЕ_mean', 'ГОД_mean',
             'СЕМЕСТР_mean', "АТДИ_mean", "КУНА_mean", "КаНА_mean", "СЕНА_mean", "ПоАТ_mean", "ПоНА_mean", "ПоДИ_mean", "ПоДИАТ_mean",
             'КУРС_mean', 'Категория обучения_mean', 'Пол_mean', 'Что именно закончил_mean', 'НАПРАВЛЕНИЕ_clustered_mean']
```

```python
X_train
```

```python
params = {
    'n_estimators': 5000,
    'depth': 10,
    'random_state':42,
    'verbose': 100,
    'eval_metric': "R2"
}
```

# Так как оценки - это условно линейная величина(т.е. можно упорядочить классы по возрастанию), можно использовать регрессор

```python
cb = catboost.CatBoostRegressor(**params)
```

```python
cb.fit(X_train[train_col], y_train)
```

# Считаем метрики

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
mean_absolute_error(y_test, cb.predict(X_test[train_col])), r2_score(y_test, cb.predict(X_test[train_col]))
```

```python
cb.get_feature_importance()
```

# Формируем ответ

```python
y_pred = pd.read_csv("sample_submission.csv", index_col=0)
y_pred["mark"] = cb.predict(X_val[train_col])
```

```python
submission = round(y_pred*2 + 3)
```

```python
submission
```

```python
import time
import os

current_timestamp = int(time.time())
submission_path = 'submissions/{}.csv'.format(current_timestamp)

if not os.path.exists('submissions'):
    os.makedirs('submissions')

submission.to_csv(submission_path, index=True)
print(current_timestamp)
```
